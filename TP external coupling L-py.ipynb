# Exemple: Exercice 3 bis

* Objectif: Simuler le développement d'une tige (exercice 3 du TD), mais depuis un script python

    * Le l-système génére le développement d'une tige dont les phytomères ne comportent que des entre-noeuds.
    * Découvrir les fonction de lpy: Lsystem, Derive
    * Editer l'axiom d'un objet lsystem
        


```python
# lecture objet lsystem
nomlsystem = 'Exo 3- Axis0.lpy'
lsys = lpy.Lsystem(nomlsystem)  # objet l-system

# lecture de l'axiom et du nombre d'iteration dans l'objet lsystem
lstring = lsys.axiom
nb_iter = lsys.derivationLength 

print('axiom :', lstring)
print('derivationLength :', nb_iter)
```

```python
# run du lsystem initial
lsys.derive()

```

```python
# run du lsystem en changeant derivationLength
nb_iter = 3
lsys.derive(nb_iter)
```

```python
# changement de l'axiom par '[A][A]A' figurant la présence de bourgeons lateraux

# cree un Axial Tree (pas une simple chaine de caracteres)
a = lpy.AxialTree() #vide
a.append('[A][A]A')

lsys.axiom = a  # passe l'axial tree comme axiom

# run du lsystem edite
lsys.derive()
```

```python

#print du l-system modifié
print(lsys) #visualisation du contenu (texte) du fichier l-system 

# visualisation LsystemWidget
LsystemWidget(nomlsystem, unit='cm',size_world=.4, animate=True)

#rq : visualise fichier non-mofifie... car prend nom du lsystem en entree...
```

# Exercice 9 bis

* Objectif: Rejouer l'exercice 9 (couplage caribu) avec un couplage externe des caclcul de rayonnement dans un script python
    * comparer les sorties avec un couplage interne / externe de la version offrant cette option ('Exo 9 - Axis2_ombre_light_caribu_external.lpy')
    * initialiser un lsystem en externe depuis python
    * faire un run step by step pour :
        - 1) récupérer à chaque itération les calcul d'environnement et 
        - 2) les réinjeter dans le l-system



```python
# Run de l'exo 9 (couplage caribu) avec un couplage interne a lpy avec le module de lumiere, dans le EndEach

nomlsystem = 'Exo 9 - Axis2_ombre_light_caribu_external.lpy' 
lsys = lpy.Lsystem(nomlsystem)
#print(lsys)

# simulation avec couplage lumiere
#active le calcule interne dans EndEach / desactive external coupling
lsys.opt_external_coupling = 0

LsystemWidget(nomlsystem, unit='cm',size_world=.4, animate=True)

```

```python
# Run de l'exo 9 (couplage caribu) avec un couplage externe dans 
from alinea.caribu.CaribuScene import CaribuScene

nomlsystem = 'Exo 9 - Axis2_ombre_light_caribu_external.lpy' 
lsys = lpy.Lsystem(nomlsystem)


def update_light_lstring(lstring, Dico_Par):
    "update de lstring avec resultats calcul lumiere ; renvoie liste des par"
    ls_par = []
    for mod in range(len(lstring)) :
        if lstring[mod].name == 'A' or lstring[mod].name == 'B':
            #index = mod.name.index
            calcpar = Dico_Par[mod]
            lstring[mod][1] = calcpar
            ls_par.append(calcpar)
    
    return lstring, ls_par


#desactive le calcule interne dans EndEach / active external coupling
lsys.opt_external_coupling = 1


# initialisation
lstring = lsys.axiom #initialise lstring avex l'axiom du l-system
nb_iter = 400

#boucle pour un calcul step bu step
for i in range(nb_iter+1):
        #print('iter ',i)
        lstring = lsys.derive(lstring, i, 1)
        lscene = lsys.sceneInterpretation(lstring)
        
        ############
        # step light transfer coupling
        ############
        
        # Calculs de caribu et sortie du Dico contenant les valeurs de rayonnement
        cc_scene = CaribuScene(scene = lscene)
        #lsys.plot(lstring)
        aggregated_out = cc_scene.run(direct = True, infinite = False)
        Dico_Par = aggregated_out[1]['default_band']['Ei'] 
        
        #mise a jour de la lstring
        lstring, lsys.lspar = update_light_lstring(lstring, Dico_Par)

        
```
